```python
from IPython.display import HTML, display

def display_table(data):
    html = "<table>"
    for row in data:
        html += "<tr>"
        for field in row:
            html += "<td>%s</td>"%(field)
        html += "</tr>"
    html += "</table>"
    display(HTML(html))
    
    
def copy_html(data):
    html = "<table>"
    for row in data:
        html += "<tr>"
        for field in row:
            html += "<td>%s</td>"%(field)
        html += "</tr>"
    html += "</table>"
    return(HTML(html))



    
```

```python
htmlTableDict = {}
```

```python
data = [[""]+funders]



for i in range(0,len(fndDf[funders[0]]['citation_stats'])-1):
    row = [fndDf[funders[0]]['citation_stats'][i]['stat']]
    for f in funders:
        val = fndDf[f]['citation_stats'][i]['value']
        if "uncited_pct" in row:
            val = str(round(100*val,1))+"%"
        row.append(val)
    data.append(row)
        
htmlTableDict['citation_stats'] = copy_html(data)
display_table(data)
```

```python

ranks = ['authors_top25','groups_top25','journals_20','top_cofunders']
```

```python
for f in funders:
    for r in ranks:
        fndDf[f][r] = [x for x in fndDf[f][r] if x[list(x.keys())[0]] != ""]
```

```python
fndDf[funders[0]][ranks[0]][0]
```

```python
for kpi in ranks:
    data = [[""]+funders]
    
    for i in range(0,len(fndDf[funders[0]][kpi])):
        
        
        
        if kpi == "top_cofunders" and i == 0:
            continue
        #print(i)
        row = [i]
        for f in funders:
            try:
                vals = list(fndDf[f][kpi][i].values())
                    
                row.append(vals[0]+" ("+str(vals[1])+")")
            except:
                row.append("")
        data.append(row)

    print('\n'+kpi)
    htmlTableDict[kpi] = copy_html(data)
    display_table(data)
```

```python
doctypes = []
for f in funders:
    dt = [list(x.values())[0] for x in fndDf[f]['doc_types']]
    for x in dt:
        doctypes.append(dt)
```

```python
doctypes = list(set(chain(*doctypes)))
```

```python
doctypes = list(set([x.strip() for x in doctypes]))
```

```python
doctypes
```

```python

    
    
    
data = [[""]+funders]



for d in doctypes:
    row = [d]
    for f in funders:
        val = [x for x in fndDf[f]['doc_types'] if d in list(x.values())]
        if len(val) == 0:
            row.append("")
        else:
            row.append(val[0]['value'])
    data.append(row)
    
        
htmlTableDict['doc_types'] = copy_html(data)     
display_table(data)
```
